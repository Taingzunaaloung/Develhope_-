# Data Cleaning Guide

# Data Cleaning

## 1.0 Drop Duplicates by 'App'

```python
df_copy = df_copy.drop_duplicates(subset=['App'], keep='first').reset_index()
df_copy
```

```python
# No duplicates in 'df_copy' dataset based on app name
duplicate_name_null = df_copy[df_copy.duplicated(subset=['App'], keep=False)].sort_values(by='App')
duplicate_name_null
```

```python
df_copy.info()
```

## 2.0 Columns Cleaning

### Column 'Size'

```python
# Find all Unique values in the column
df_copy['Size'].unique()
```

```python
# Row with 'Size' = '1,000+'
row = df_copy[df_copy['Size'] == '1,000+']
row
```

```python
# Shift the row one place to the right starting from the column 'Category'(index: column=2, row=9330)

# save the row to be corrected in a variable
row_to_correct = df_copy.loc[9300]

# shift the values to the right starting from the second column using .shift() method
row_to_correct.iloc[2:] = row_to_correct.iloc[2:].shift(1)

# update the original row with the corrected row
df_copy.loc[9300] = row_to_correct

df_copy.loc[9300]
```

```python
# Convert 'Size' column to string
df_copy['Size'] = df_copy['Size'].astype(str)
df_copy.info()
```

```python
# Split string into numeric value and unit of measurement in 2 new columns 'Size_numeric' and 'Size_unit'
df_copy['Size_numeric'] = df_copy['Size'].str.extract('(\d+\.?\d*)', expand=False).astype(float)
df_copy['Size_unit'] = df_copy['Size'].str.extract('([a-zA-Z]+)', expand=False)
df_copy
```

```python
# Convert rows with size in megabytes to kilobytes
df_copy.loc[df_copy['Size_unit'] == 'M', 'Size_numeric'] *= 1024

# Fill empty rows in 'Size_numeric' column with the average size of apps in the same category
df_copy.loc[df_copy['Size_unit'] == 'Varies', 'Size_numeric'] = df_copy.groupby('Category')['Size_numeric'].transform('mean').astype(float).round(1)

# Drop 'Size_unit' column
df_copy = df_copy.drop(['Size_unit'], axis=1)

# Replace 'Size'values with 'Size_numeric' values
df_copy['Size'] = df_copy['Size_numeric']

# Drop 'Size_numeric' column
df_copy = df_copy.drop(['Size_numeric'], axis=1)
df_copy
```

```python
df_copy.info()
```

### 2.2 Column 'Installs'

```python
# Find all unique values in the column to ensure they can be converted to numeric
df_copy['Installs'].unique()

# For the column 'Installs', the non numeric characters are ','and '+'
```

```python
# Remove all ',' and '+' from column 'Installs'
df_copy['Installs'] = df_copy['Installs'].str.replace(',', '', regex=False).str.replace('+', '', regex=False)

# Convert the column from object to numeric
df_copy['Installs'] =pd.to_numeric(df_copy['Installs'])
```

```python
df_copy.info()
```

```python
df_copy
```

### 2.3 Column 'Type'

```python
# Show unique values in the column
df_copy['Type'].unique()
```

```python
# Identify row with null value in column 'Type'
null = df_copy[df_copy['Type'].isnull()]
null
```

```python
# Fill the null with 'Free' since the price is 0
df_copy.loc[8028, 'Type'] = 'Free'
df_copy.loc[8028]
```

```python
df_copy['Type'].unique()
```

### 2.4 Column 'Price'

```python
df_copy['Price'].unique()
```

```python
# Remove the dollar sign and trailing spaces from the Price column and convert to float
df_copy['Price'] = df_copy['Price'].str.replace('$', '').astype(float)
df_copy
```

```python
df_copy.info()
```

### 2.5 Column 'Category'

```python
df_copy['Category'].unique()
```

```python
# Find cell with value null
null_cat = df_copy[df_copy['Category'].isnull()]
null_cat
```

```python
# Fill empty cell with 'Unknown'
df_copy.loc[9300, 'Category'] = 'Unknown'
df_copy.loc[9300]
```

### 2.6 Column 'Rating'

```python
# Find all Unique values in the column
df_copy['Rating'].unique()
```

```python
# Find cells with null value in 'Rating' 
null_rating = df_copy[df_copy['Rating'].isnull()]
null_rating
```

```python
df_copy.info()
```

```python
# Fill all null cells with the average rating of all apps in the same category
df_copy['Rating'] = df_copy.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Convert 'Rating' dtype to float and round to one decimal place
df_copy['Rating'] = df_copy['Rating'].astype(float).round(1)
```

```python
df_copy.info()
```

# Top 10 Most Installed Apps

```python
# Top 10 most installed apps irrespective of repeating install number 
top_10_installed1 = df_copy.sort_values(by='Installs', ascending=False).head(10)
plt.bar(top_10_installed1['App'], top_10_installed1['Installs'])
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Top 10 most installed apps, only taking the first of each 'Installs' band
top_101 = df_copy.sort_values(by=['Installs', 'App'], ascending=[False, True]).drop_duplicates(subset=['Installs']).head(10)
plt.bar(top_101['App'], top_101['Installs'])
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Most Installed Categories

```python
popular_category = df_copy.groupby('Category')['Installs'].mean().sort_values(ascending=False)
# create a bar chart to visualize the data
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(popular_category.index, popular_category.values)
# to make the chart horizontal, use ax.barh instead of ax.bar above
ax.set_xlabel('Categories')
ax.set_title('Average number of installs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Paid Apps with Biggest Revenue

```python
# Paid apps
paid_apps = df_copy[df_copy['Type'] =='Paid']

# Create a new column for revenue for 'paid_apps' by multiplying the price column with the number of installs
paid_apps['Revenue'] = paid_apps['Price'] * paid_apps['Installs']

# sort paid apps by revenue in desc. order. the top 'n' apps with biggest revenue can be found by using .head(n)
paid_apps.sort_values(by='Revenue', ascending=False)
```

```python
to_group = paid_apps.sort_values(by='Revenue', ascending=False).head(100)
top_rev_cat = to_group.groupby('Category').count().sort_values(by='App', ascending=False)
top_rev_cat
```
